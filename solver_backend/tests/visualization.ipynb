```python
import matplotlib.pyplot as plt
import json
import numpy as np

with open("all_solvers_data_5.json", "r") as file:
    data = json.load(file)
```

```python
dims = data.keys()
solvers = ["genetic", "dynamic", "order", "brute_force", "minizinc"]
res = {}

for solver in solvers:
    res[solver] = {}
    res[solver]["cost"] = [data[dim][solver]["cost"] for dim in dims if solver in data[dim] and data[dim][solver]]
    res[solver]["time"] = [data[dim][solver]["time"] for dim in dims if solver in data[dim] and data[dim][solver]]


# minizinc_cost = [data[dim]["minizinc"]["cost"] for dim in dims]
# minizinc_time = [data[dim]["minizinc"]["time"] for dim in dims]
# 
# genetic_cost = [data[dim]["genetic"]["cost"] for dim in dims]
# genetic_time = [data[dim]["genetic"]["time"] for dim in dims]
```

```python
plt.plot([i for i in range(len(res["genetic"]["cost"]))], res["genetic"]["cost"], label="genetic")

plt.plot([i for i in range(len(res["order"]["cost"]))], res["order"]["cost"], label="order")

plt.plot([i for i in range(len(res["brute_force"]["cost"]))], res["brute_force"]["cost"], label="brute_force")

plt.plot([i for i in range(len(res["minizinc"]["cost"]))], res["minizinc"]["cost"], label="minizinc")

plt.plot([i for i in range(len(res["dynamic"]["cost"]))], res["dynamic"]["cost"], label="dynamic")

plt.yscale('log')

plt.title("Cost over dimensions")

plt.xlabel("Matrix dimensions")
plt.ylabel("Solution Cost")
plt.legend()
plt.savefig("cost_plot.jpg")
```

```python
plt.plot([i for i in range(len(res["genetic"]["time"]))], res["genetic"]["time"], label="genetic")

plt.plot([i for i in range(len(res["order"]["time"]))], res["order"]["time"], label="order")

plt.plot([i for i in range(len(res["brute_force"]["time"]))], res["brute_force"]["time"], label="brute_force")

plt.plot([i for i in range(len(res["dynamic"]["time"]))], res["dynamic"]["time"], label="dynamic")

plt.plot([i for i in range(len(res["minizinc"]["time"]))], res["minizinc"]["time"], label="minizinc")

plt.yscale('log')

plt.title("Time over dimensions")

plt.xlabel("Matrix dimensions")
plt.ylabel("Solution Time")
plt.legend()
plt.savefig("time_plot.jpg")
```
